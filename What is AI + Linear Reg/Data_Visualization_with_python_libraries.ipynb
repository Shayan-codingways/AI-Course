# **Python Libraries for Data Visualization**

## **Matplotlib – The Classic Visualizer**

Matplotlib is imported to create visual plots like line graphs, bar charts, etc.

x and y lists are defined to store the data for plotting.

**plt.plot()** generates the line graph, plotting x values on the horizontal axis and y values on the vertical axis.

**plt.xlabel()** and **plt.ylabel()** add labels to the respective axes for clarity.

**plt.title()** sets the title of the plot.

**plt.legend()** displays a legend to label the graph.

Finally, **plt.show() **renders and displays the plot on the screen.



```python
# Import the Matplotlib library for plotting
import matplotlib.pyplot as plt

# Define x-axis data points
x = [1, 2, 3, 4, 5]

# Define y-axis data points (square of x values)
y = [1, 4, 9, 16, 25]

# Create a line plot using x and y data
plt.plot(x, y, label="Square Numbers")

# Add a label for the x-axis
plt.xlabel('X-axis')

# Add a label for the y-axis
plt.ylabel('Y-axis')

# Add a title to the plot
plt.title('Simple Line Plot')

# Display the legend to show the label
plt.legend()

# Display the plot on the screen
plt.show()
```

### **Bar chart**

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

# Creating a simple bar chart
plt.figure(figsize=(4, 5))
plt.bar(categories, values, color='orange')
plt.title('Simple Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

## **Seaborn – Statistical Beauty**

Seaborn is a statistical visualization library built on Matplotlib.

**Data Preparation:**

**pd.DataFrame({...})**: Creates a DataFrame with 8 columns (variables) named 'A' through 'H' and 8 rows of sample data.
Pair Plot Creation:

**sns.pairplot(data)**: Generates a pair plot to visualize the pairwise relationships among all 8 variables in the DataFrame. Each cell in the plot shows the scatter plot of two variables, and the diagonal shows the distribution of each variable.
Plot Customization:

**plt.suptitle('Pair Plot of 8 Variables', y=1.02**): Adds a title to the pair plot with a slight adjustment to its position (y=1.02) to ensure it does not overlap with the plot.
Display:

**plt.show()**: Displays the pair plot.

```python
# Import Seaborn, Matplotlib, and Pandas
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a simple DataFrame with 8 variables (features)
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5, 6, 7, 8],
    'B': [8, 7, 6, 5, 4, 3, 2, 1],
    'C': [1, 3, 2, 4, 6, 5, 7, 8],
    'D': [5, 3, 6, 4, 7, 8, 2, 1],
    'E': [8, 5, 6, 3, 2, 7, 1, 4]
})

# Create a pair plot using Seaborn
sns.pairplot(data)

# Add a title to the pair plot
plt.suptitle('Pair Plot of 8 Variables', y=1.02)

# Display the plot
plt.show()

```

## ** Plotly – Interactivity at Its Best**

Plotly Express is imported to create interactive plots.

**px.data.iris()** loads the famous Iris flower dataset.

**px.scatter() **creates an interactive scatter plot, using sepal_width and sepal_length for the x and y axes.

**color='species'**  colors the points based on different flower species.

**fig.show()** renders the plot with interactive features like zooming and hovering.

```python
# Import the Plotly Express module for interactive plots
import plotly.express as px

# Load a built-in dataset ('iris') from Plotly
df = px.data.iris()

# Create an interactive scatter plot
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species', title='Interactive Scatter Plot')

# Display the interactive plot
fig.show()
```

## **Pandas Visualization**

Pandas is imported for data manipulation and visualization.

x and y lists define the data points.

A DataFrame is created from the lists, making it easier to work with the data.

**data.plot()** generates a line plot directly from the DataFrame.

**plt.show(**) renders the plot on the screen using Matplotlib.

```python
# Import Pandas for data manipulation and plotting
import pandas as pd

# Define x and y data points
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create a DataFrame from the data
data = pd.DataFrame({'x': x, 'y': y})

# Use the built-in plot function to create a line plot
data.plot(x='x', y='y', kind='line', title='Pandas Line Plot')

# Show the plot
plt.show()
```

## **ggplot (plotnine)**

plotnine is a Python implementation of ggplot2, designed for grammar-based visualizations.

x and y data are simple lists of numbers.

A DataFrame is created from the lists to feed into ggplot.

**ggplot()** creates the plot, and **aes()** specifies the data mappings for the axes.

**geom_line()** adds a line geometry to the plot.

The **print()** function displays the plot in the notebook.

```python
# Import ggplot from plotnine for visualizations
from plotnine import ggplot, aes, geom_line
import pandas as pd

# Define x and y values as simple lists
x = [1, 2, 3, 4, 5]
y = [3, 7, 9, 14, 20]

# Create a DataFrame from the data
data = pd.DataFrame({'x': x, 'y': y})

# Create a ggplot chart with aesthetics (x, y) and a line geometry
plot = ggplot(data, aes(x='x', y='y')) + geom_line()

# Display the plot
print(plot)
```

## **Libraries and Financial Data**

```python
# Import Seaborn, Matplotlib, and Pandas
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame with financial-related data (e.g., stock prices, returns, P/E ratio, market cap, etc.)
data = pd.DataFrame({
    'Stock Price': [120, 125, 130, 128, 132, 135, 140, 145],
    'P/E Ratio': [25, 30, 28, 32, 35, 37, 38, 40],
    'Market Cap (B)': [300, 320, 310, 315, 330, 340, 350, 360],
    'Return (%)': [2, 2.5, 3, -1.5, 1, 3.5, 4, 2.5],
    'Volume (M)': [10, 12, 8, 7, 9, 11, 13, 14]
})

# Create a pair plot using Seaborn for financial features
sns.pairplot(data)

# Add a title to the pair plot
plt.suptitle('Pair Plot of Financial Variables', y=1.02)

# Display the plot
plt.show()
```

```python
# Import necessary libraries
import yfinance as yf  # For downloading stock data from Yahoo Finance
import plotly.graph_objects as go  # For creating candlestick charts with Plotly

# Download historical stock data for Apple ('AAPL') using yfinance
# The data will include Open, High, Low, Close, Volume, and other columns
# We are downloading data from January 1, 2023, to January 30, 2023
data = yf.download('AAPL', start='2023-01-01', end='2023-01-30')

# Create a candlestick chart using the real stock data we downloaded
# 'x' is the date (index), and we provide 'Open', 'High', 'Low', and 'Close' prices
fig = go.Figure(data=[go.Candlestick(x=data.index,              # Date (x-axis)
                                     open=data['Open'],         # Opening prices
                                     high=data['High'],         # Highest prices
                                     low=data['Low'],           # Lowest prices
                                     close=data['Close'],       # Closing prices
                                     increasing_line_color='green',  # Green for increasing candles
                                     decreasing_line_color='red')])  # Red for decreasing candles

# Update layout to include title and other layout features
fig.update_layout(title='Apple Stock Price Candlestick Chart')  # Set chart title

# Display the candlestick chart
fig.show()  # Renders the chart in the notebook or IDE
```

Matplotlib

```python
# Import necessary libraries
import yfinance as yf  # For downloading stock data
import matplotlib.pyplot as plt  # For plotting the data

# Download historical stock data for Apple (AAPL) using yfinance
# Data will include Open, High, Low, Close prices, and Volume
data = yf.download('AAPL', start='2023-01-01', end='2023-01-30')

# Calculate a simple moving average (SMA) for a 7-day period
data['SMA_7'] = data['Close'].rolling(window=7).mean()

# Plotting stock price (Close) and moving average using Matplotlib

# Create the plot
plt.figure(figsize=(10,6))  # Set the figure size

# Plot the closing prices
plt.plot(data.index, data['Close'], label='Apple Close Price', color='blue', linewidth=2)

# Plot the 7-day moving average
plt.plot(data.index, data['SMA_7'], label='7-Day Moving Average', color='orange', linewidth=2)

# Set the chart title and labels
plt.title('Apple Stock Price with 7-Day Moving Average (Jan 2023)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)

# Show the legend
plt.legend(loc='upper left')

# Display the plot
plt.grid(True)  # Add grid lines for better readability
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Automatically adjust subplot parameters for better display

# Show the final plot
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import yfinance as yf  # For downloading stock data
import mplfinance as mpf  # For creating candlestick charts using Matplotlib

# Download stock data (e.g., Apple) using yfinance
# The data will include Open, High, Low, Close, Volume, etc.
data = yf.download('AAPL', start='2023-01-01', end='2023-01-30')

# Set the index of the DataFrame to be the date, if not already
data.index.name = 'Date'

# Use mplfinance to create a candlestick chart
# type='candle' specifies that we are creating a candlestick chart
mpf.plot(data, type='candle', style='charles',
         title='Apple Stock Price Candlestick Chart (January 2023)',
         ylabel='Price (USD)',
         volume=True,  # Show the volume at the bottom
         mav=(5,10),  # Add moving averages (5 and 10 days)
         show_nontrading=False)  # Hide non-trading days
```

```python
import seaborn as sns
import pandas as pd

# Load the stock data
url = 'https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv'
df = pd.read_csv(url)

# Extract month from the Date column
df['Month'] = pd.to_datetime(df['Date']).dt.month

# Create a box plot of total funds by month
plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Total', data=df)
plt.title('Monthly Distribution of Total Funds')
plt.xlabel('Month')
plt.ylabel('Total Funds')
plt.show()
```
