## **KDE PLOT**

It is a way to estime probablity density function of a continous random function

distplot(data) is used to visualize parametric distribution of data.

```python
sns.distplot(full_data['Age'], hist=False)
plt.show
# we can see most of the passenger has age bw 20/40
```

```python
plt.figure(figsize=(8,8))
sns.kdeplot(full_data['Age'], shade=True)
plt.show
```

KDE stands for Kernel Density Estimation, a non-parametric way to estimate the probability density function (PDF) of a random variable. It is often used to smooth out data points and understand the distribution of a dataset.

In simpler terms, KDE helps visualize the underlying distribution of data by generating a smooth curve (density estimate) from a set of data points. It's useful in exploratory data analysis, especially when comparing how the data is distributed compared to the normal distribution or other theoretical distributions.




The distplot function, from the Seaborn library in Python, is used to visualize the distribution of a dataset. It combines a histogram and a Kernel Density Estimate (KDE) plot into one, allowing you to observe both the frequency of data points and an estimate of the probability density function.

Key features of distplot:

Histogram: Shows the count of observations within different bins or intervals.
KDE Plot: A smoothed line that represents the distribution of the data.
Rug Plot (optional): Small vertical lines at the base of the plot representing individual data points.
Since Seaborn version 0.11.0, distplot has been deprecated, and its functionality has been split into two separate functions:

histplot: For creating histograms.
kdeplot: For generating kernel density plots.

## **Categorical** **Plots**

```python
sns.barplot( x='Pclass', y='Fare' , hue='Sex', data=full_data)
plt.show
# barplots are used for atleast categorical data and histograms are used for numeric data
```

```python
sns.stripplot( x='Sex', y='Age' , hue='Sex', data=full_data)
plt.show

# when one of vaiable is categorical
```

```python
sns.boxplot( x='Survived', y='Age' , hue='Sex', data=full_data)
plt.show

# 1 data categorical
```

```python
sns.violinplot( x='Survived', y='Age' , hue='Sex', data=full_data)
plt.show

# combination of box plot and kde --> wides region, higher kde
# has box plot in between and rotated plots
#Key features of a violin plot:
#Density Shape: The width of the violin plot represents the density of the data at different values. The fatter sections of the plot indicate where more data points are concentrated.
#Symmetry: The plot is usually symmetric along its vertical axis, giving a clear picture of how data is distributed around the median.
#Box Plot (Optional): Inside the violin, there is often a mini box plot that shows summary statistics like the median, interquartile range (IQR), and outliers.
#Multiple Distributions: You can compare the distributions of different groups or categories side by side.
```
