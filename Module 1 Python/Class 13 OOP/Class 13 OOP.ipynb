# Redefine Student

```python
class Students():
    
    # initializer/constructor --> special method
    def __init__(self , name , age , city , course):
        self.name = name
        self.age = age
        self.city = city
        self.course = course

    # methods (functions within a class are called methods)
    def appear_in_exam(self):
        print(f"{self.name} is appearing in exam.")

    def pay_fee(self):
        print(f"{self.name} is paying fee.")
```

```python
s4 = Students()  

# bcz we have initializer thus, arguments missing. 
# initializer is auto called when object is called. 

# By default we have initializer and takes nothing. 
```

```python
s4 = Students('Shayan' , 22 , "Karachi" , 'Python')  # mandatory attributes
```

```python
s5 = Students('Uswa' , 20 , "Karachi" , 'Python')
```

```python
s6 = Students('Uswa' , 20 , "Karachi")
```

```python
s4.appear_in_exam()
```

```python
s5.appear_in_exam()
```

```python
class Student():
    
    # initializer/constructor --> special method
    def __init__(self , name , age , city , course = 'Python'):
        self.name = name
        self.age = age
        self.city = city
        self.course = course

    # methods (functions within a class are called methods)
    def appear_in_exam(self):
        print(f"{self.name} is appearing in exam.")

    def pay_fee(self):
        print(f"{self.name} is paying fee.")
```

```python
s11 = Student("Shayan",22,"Karachi")  # 4th argument by default
```

```python
s11.course
```

# Word --> | **Self** |

```python
class Car():
    def __init__(self, model,make,color):
        self.make = make
        self.model = model
        self.color = color

    def describe_car(self):
        print(f"""
               Model Name: {self.model}
               Car Make: {self.make}
               Car Color: {self.color}
               """)
```

```python
# self itself is an object 
```

```python
c1 = Car("2024","Toyota","Black")
c2 = Car("2025","Honda","white")
```

```python
# 3 values we gave went in the parameters, and in self -->  reference of object c1 goes. 
```

```python
# self.name = c1.name 
# self.name = c2.name
```

```python
c1.describe_car()  # self refrences specific copy of c1
```

```python
c2.describe_car()  # self refrences specific copy of c2
```

```python
Car.describe_car() # won't work as it doesn't know from which copy it should bring info from
```

```python
Car.describe_car(c1)
```

```python
Car.describe_car(c2)
```

## You can also change self to anyother name like selfie, but it should be the first parameter in initializer.

```python
class Cars():
    def __init__(selfee, model,make,color):
        selfee.make = make
        selfee.model = model
        selfee.color = color
        selfee.ac = "Dawlence" # hardcore!

    def describe_car(selfee):
        print(f"""
               Model Name: {selfee.model}
               Car Make: {selfee.make}
               Car Color: {selfee.color}
               Car AC: {selfee.ac}
               """)
```

```python
c1=Cars(2023,"Tysd", "green")
```
