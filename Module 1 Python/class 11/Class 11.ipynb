## Different Excepts for different task

```python
try:
   num1 = int(input("Enter a number - numerator: "))
   num2 = int(input("Enter a number - denominator: "))
   print(num1/num2) # --> will throw an error here. 
except ZeroDivisionError:
    print("Insan Banja mana karne kai bawajud 0 daldya") 
    
# this exception won't work as try must be on a number input. 
except ValueError:
    print("Angraiz, Number Dalna hai") 
```

### Multiple errors in one except

```python
try:
   num1 = int(input("Enter a number - numerator: "))
   num2 = int(input("Enter a number - denominator: "))
   print(num1/num2) # --> will throw an error here. 
except (ZeroDivisionError, ValueError):
    print("Insan Banja mana karne kai bawajud 0 daldya") 
```

```python
try:
    num1 = int(input("Enter a number "))
    num2 = int(input("Enter another number "))
    y=num1/num2
    
except Exception as e:
  print(f"Exception generated and Caught '{e}'")
    
# Print only if the error is not thrown/ error is not caught by the exception block. 
# so 'y not defined' won't be thrown.  
else:
  print(y)
```
