# Context Manager

# Readline()

```python
with open("myfile","r") as f:
    content = f.readline()
    print(content)
```

```python
with open("myfile","r") as f:   
    content = f.readlines()
    print(content)
# readlines returns a list with all lines, just like a list of lines we gave in writelines(). 
```

```python
with open("myfile","r") as f:
  for line in f:
      print(line)
```

# r+ mode does both read and write but doesn't create a new file. 

```python
try:
    
    with open("functions.txt", "r+") as f:
        f.write("functions are a block of code!")
        print("File Written")
        f.read()
        print("File Read______")
    
except FileNotFoundError:
    print("File not Found")
 
```

```python
try:
    
    with open("functions.txt", "r+") as f:
        f.write("functions are a block of code!")
        print("File Written")
        f.read()
        print("File Read______")
    
except Exception as e:
    print(e)
```

# w+ mode read and writes both and also creates a file. 

```python
try:
    
    with open("functions.txt", "w+") as f:
        f.write("functions are a block of code!")
        print("File Written")
        print(f.read())
        print("File Read______")
    
except FileNotFoundError:
    print("File not Found")
```

```python
# issue here it isn't reading the context as after writing the cursor is at the end and therefore reads nothing. 
```

```python
try:
    
    with open("functions.txt", "w+") as f:
        f.write("functions are a block of code!")
        print("File Written")
        f.seek(0) # Move to index 0 to read again
        print(f.read())
        print("File Read______")
    
except FileNotFoundError:
    print("File not Found")
```
