### ðŸ”´ Underfitting:

* **Definition**: When a model is too simple to capture the underlying pattern in the data.
* **Symptoms**: Poor performance on both training and testing data.
* **Example**: Using a straight line (linear regression) to fit data that has a curve or more complexity.
* **Cause**: Model lacks the capacity (too few features or too simple a function).

---

### ðŸ”µ Overfitting:

* **Definition**: When a model is too complex and starts learning the noise along with the signal in the training data.
* **Symptoms**: Very high accuracy on training data, but poor performance on unseen test data.
* **Example**: A high-degree polynomial fitting every data point exactly.
* **Cause**: Too many features, not enough training data, or not enough regularization.

---

### ðŸŸ¢ Best Fit (Good Fit):

* **Definition**: A model that captures the underlying trend in the data without memorizing the noise.
* **Symptoms**: Good performance on both training and test/validation data.
* **Example**: A balanced model that generalizes well to new, unseen data.
* **Goal**: Achieve the lowest possible error on unseen data (good **generalization**).

---

### Visual Summary:

If you were to visualize this using a curve fitting points on a graph:

* **Underfit**: The curve is too simple, missing trends.
* **Overfit**: The curve is too wiggly, chasing every point.
* **Best Fit**: The curve smoothly follows the overall pattern.




### ðŸ”· **Bias**

* **Definition**: Error due to overly simplistic assumptions in the model.
* **High Bias** â†’ Model **misses important patterns** in the data (underfitting).
* **Low Bias** â†’ Model captures more of the dataâ€™s complexity.

**Example**: Using a straight line (linear model) to fit a curved relationship.

---

### ðŸ”¶ **Variance**

* **Definition**: Error due to **too much sensitivity to small fluctuations** in the training data.
* **High Variance** â†’ Model learns **noise**, not just the pattern (overfitting).
* **Low Variance** â†’ Model is more stable across different datasets.

**Example**: Using a very complex model (like a 10th-degree polynomial) that fits training data exactly but performs poorly on new data.

---

### âš–ï¸ Bias-Variance Tradeoff

|                  | Bias                               | Variance |
| ---------------- | ---------------------------------- | -------- |
| **Underfitting** | High                               | high      |
| **Overfitting**  | Low                                | High     |
| **Best Fit**     | Balanced (Low bias + Low variance) |          |

The goal in machine learning is to **find the sweet spot** between bias and variance â€” **a model that generalizes well**.

---

### ðŸŽ¯ Analogy (Easy to Remember):

Imagine shooting arrows at a target:

* ðŸŽ¯ **High Bias, Low Variance**: All arrows are grouped together but far from the bullseye.
* ðŸŽ¯ **Low Bias, High Variance**: Arrows are scattered widely but centered around the bullseye.
* ðŸŽ¯ **Low Bias, Low Variance**: Arrows are tightly grouped and close to the bullseye â€” this is the ideal.

# Ridge Regression

```python
# This model doesn't fits well as seen, linear regression isn't good. 
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Model 
ridge = Ridge(alpha = 10) # alpha is penalty for regularization

# Model Train
ridge.fit(X_train,y_train)
```

```python
# Model prediction
y_train_pred = ridge.predict(X_train)
print(mean_squared_error(y_train, y_train_pred))
print(r2_score(y_train, y_train_pred))
```

```python
y_test_pred = ridge.predict(X_test)
print(mean_squared_error(y_test, y_test_pred))
print(r2_score(y_test, y_test_pred))
```

# Grid Search

```python
# Grid Search gives the best alpha value for best results 

# Hyperparameters are the settings you choose before training a machine learning model â€” theyâ€™re not learned from the data, but manually set or tuned to control the training process.

# Hyperparameters are knobs you turn to shape how your model learns.

from sklearn.model_selection import GridSearchCV

# List of hyperparamters
param_grid = {'alpha': [0.0001,0.001,0.01,0.1,1,10,100,1000]}

# model ridge with grid search
ridge_cv = GridSearchCV(ridge, param_grid = param_grid, cv = 2) # --> cv (cross validation)
# train model
ridge_cv.fit(X_train, y_train)


y_pred_train = ridge_cv.predict(X_train)

print(mean_absolute_error(y_train,y_pred_train))
print(mean_squared_error(y_train,y_pred_train))
print(r2_score(y_train,y_pred_train))
print(ridge_cv.best_estimator_)
```

```python
y_pred_test = ridge_cv.predict(X_test)

print(mean_absolute_error(y_test,y_pred_test))
print(mean_squared_error(y_test,y_pred_test))
print(r2_score(y_test,y_pred_test))
print(ridge_cv.best_estimator_)
```

# ðŸ•’ 1. Temporal Leak (also called Data Leakage in Time Series)
ðŸ” Definition:
Temporal leak happens when future data is accidentally used to train a model that is supposed to make predictions about the future. This violates the time-order of the data, leading to unrealistically good performance during model evaluation.

âŒ Why It's a Problem:
The model "cheats" by learning from information it wouldn't have access to in a real-world future prediction scenario. This causes overfitting and poor generalization.

âš ï¸ Example:
Suppose you're building a stock price prediction model using historical data from 2010 to 2020.

âœ… Correct: Train on 2010â€“2018, validate on 2019, test on 2020.

âŒ Leak: You accidentally include 2019 prices while training, even though your goal is to predict for 2019.

# ðŸ” **What is Cross-Validation?**

**Cross-validation** is a technique used to **evaluate the performance of a machine learning model** by testing it on **different subsets of the data**.
It helps ensure your model **generalizes well** to unseen data, not just the training data.

---

### ðŸ“¦ **Why Do We Use It?**

* To **avoid overfitting** (model performing well on training data but poorly on new data)
* To get a **more reliable estimate of model accuracy**
* To make the most of **limited data**

---

## ðŸ” **How It Works (Most Common Type: K-Fold Cross-Validation)**

1. **Split the dataset into â€œkâ€ equal parts** (called folds) â€” e.g., k = 5
2. **Train the model k times**, each time using:

   * kâˆ’1 folds for training
   * 1 fold for testing (a different one each time)
3. **Average the performance** across all k tests

---

### ðŸ“Š Example (5-Fold Cross-Validation):

| Fold | Training Data Used | Testing Data Used |
| ---- | ------------------ | ----------------- |
| 1    | Folds 2, 3, 4, 5   | Fold 1            |
| 2    | Folds 1, 3, 4, 5   | Fold 2            |
| 3    | Folds 1, 2, 4, 5   | Fold 3            |
| 4    | Folds 1, 2, 3, 5   | Fold 4            |
| 5    | Folds 1, 2, 3, 4   | Fold 5            |

Then, the final performance = **average of all 5 test scores**

---

### âœ… Benefits:

* Uses **all data** for both training and testing
* Gives a **more robust** measure of model accuracy
* Helps in **model selection and hyperparameter tuning**

---

### â— Tip:

Always use **cross-validation** especially when your dataset is **not very large**, to avoid misleading results.

## ðŸŽ **Imagine This: Youâ€™re Grading a Student's Presentation**

You want to be fair, so instead of grading just **once**, you do it **5 times**, each time with a different **group of judges**.

---

### ðŸ‘¥ You have:

* 5 judges (folds)
* The student presents 5 times
* Each time, **4 judges watch and give feedback**, and **1 judge just evaluates**

---

### Round-by-Round:

| Round | Presenting to Judges | Evaluated By |
| ----- | -------------------- | ------------ |
| 1     | Judge 2, 3, 4, 5     | Judge 1      |
| 2     | Judge 1, 3, 4, 5     | Judge 2      |
| 3     | Judge 1, 2, 4, 5     | Judge 3      |
| 4     | Judge 1, 2, 3, 5     | Judge 4      |
| 5     | Judge 1, 2, 3, 4     | Judge 5      |

At the end, you **average all 5 scores** to get a **fair final grade**.

---

## ðŸ§  Now Relate It to Machine Learning:

* The **student** = your **machine learning model**
* The **judges** = your **data split into 5 folds**
* You train the model 5 times on different pieces of data and test it on the rest.

---

## âœ… Final Result:

* You get a **more balanced and fair** idea of how your model performs on **unseen data**.
