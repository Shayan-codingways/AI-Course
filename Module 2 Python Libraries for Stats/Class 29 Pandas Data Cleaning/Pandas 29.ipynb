# Reloading and Rehandling Data 

```python
data = pd.read_excel("employee.xlsx")
data
```

## data.dropna(axis=0) removes rows that contain any missing (NaN) values from the DataFrame data.

Explanation:
axis=0  Specifies that the operation should be performed on rows (not columns).
It returns a new DataFrame with the affected rows removed, but does not modify the original DataFrame unless inplace=True is used.


```python
data.dropna(axis=0)
```

## data.dropna(axis=1) removes columns that contain any missing (NaN) values from the DataFrame data.

Explanation:
axis=1  This specifies that the operation should be performed on columns (not rows).
By default, dropna() removes only columns that contain at least one NaN value.
It returns a new DataFrame with those columns removed but does not modify the original DataFrame unless inplace=True is used.

```python
data
```

```python
data.dropna(axis=1)
```

## data.dropna(how="all") removes only rows where all values are NaN (missing values).

Explanation:
how="all"  Drops only rows where every column has NaN values.
It does not drop rows that contain at least one non-NaN value.
It returns a new DataFrame with the affected rows removed but does not modify the original DataFrame unless inplace=True is used.

```python
data
```

```python
data.dropna(how="all")
```

## data.dropna(how='any') removes rows that contain at least one NaN (missing) value.

Explanation:
how="any"  Drops any row that has at least one NaN value.
It returns a new DataFrame with affected rows removed but does not modify the original DataFrame unless inplace=True is used.

```python
data.dropna(how = 'any')
```

## data.dropna(thresh=int(data.shape[0] * 0.95), axis=1) removes columns that have less than 95% non-null values.

Explanation:
data.shape[0]  Total number of rows in the DataFrame.
int(data.shape[0] * 0.95)  This calculates 95% of the total rows (rounded down).
thresh=<value>  Requires at least this many non-null values to keep a column.
axis=1  The operation is applied to columns.


```python
data.dropna(thresh=16 , axis=1) ## data.dropna(thresh=16, axis=1) removes columns that have less than 16 non-null (non-NaN) values.
```

```python
data.dropna( thresh = int(data.shape[0] * .95) , axis=1)
```

```python
data.dropna( thresh = int(data.shape[0] * 1) , axis=1)
```

## data.dropna(how="any", subset=['Emp_ID', 'Salary']) removes rows where either 'Emp_ID' or 'Salary' has a missing (NaN) value.

Explanation:
how="any"  Drops a row if at least one of the specified columns (Emp_ID, Salary) has NaN.
subset=['Emp_ID', 'Salary']  Only considers missing values in these two columns, ignoring other columns.
It returns a new DataFrame but does not modify the original unless inplace=True is used.

```python
data.dropna(how="any", subset=['Emp_ID', 'Salary'])
```

## Now doing it actually

```python
data.fillna({'Emp_ID':data.Emp_ID.mode()[0],
             "Designation":data.Designation.mode()[0],
             "Department":data.Department.mode()[0],
             "Age":data.Age.median(),
             "Salary":data.Salary.median()},inplace=True)
```
