```python
ravelled = arr.ravel()
ravelled
```

```python
ravelled
```

```python
arr
```

###### We changed in Ravelled / View but it changed our entries in our original array. This is the concept of views

In NumPy, a view is a new array object that shares the same data as the original array. This means changes made to the view will also affect the original array (and vice versa), as they are referencing the same memory location for the data.

Key Features of a View:
Shared Data: [*A view does not copy the data*#]. Instead, it points to the same data in memory as the original array.

Different Array Object: A view is a separate array object, so it can have its own shape, strides, or slicing, but the data remains the same.

Memory Efficiency: Views are more memory-efficient because they avoid duplicating data.
